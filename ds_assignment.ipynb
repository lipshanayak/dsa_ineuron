```python
# 12. Implement a function to check if a given string is an anagram of another string.
def is_anagram(str1, str2):
    
    if len(str1) != len(str2):
        return False
    
    sorted_str1 = sorted(list(str1))
    sorted_str2 = sorted(list(str2))
    
    if sorted_str1 == sorted_str2:
        return True
    
    return False

string1 = "top"
string2 = "pot"
if is_anagram(string1, string2):
    print("Anagram")
else:
    print("Not an Anagram")
```

```python
# 14. Implement a function to calculate the Fibonacci sequence up to a given number of terms.
def fibonacci_sequence(n):
    sequence = []
    if n >= 1:
        sequence.append(0)
    if n >= 2:
        sequence.append(1)
    if n >= 3:
        for i in range(2, n):
            next_term = sequence[i - 1] + sequence[i - 2]
            sequence.append(next_term)
    return sequence

terms = int(input("Enter the number of terms: "))
fibonacci = fibonacci_sequence(terms)
print("Fibonacci sequence:", fibonacci)

    
```

```python
# 18. Implement a function to find the maximum subarray sum in a given list.
def max_subarray_sum(lst):
    max_sum = np.NINF  
    current_sum = 0  
    
    for num in lst:
        current_sum += num
        
        if current_sum > max_sum:
            max_sum = current_sum
        
        if current_sum < 0:
            current_sum = 0
    
    return max_sum

my_list = np.array([-2, 1, -3, 4, -1, 2, 1, -5, 4])
max_sum = max_subarray_sum(my_list)
print("Maximum subarray sum:", max_sum)
```

```python
# 20. Implement a function to reverse the order of words in a given sentenc
def reverse_words(sen):
    words = sen.split()  
    reversed_sen = ' '.join(reversed(words)) 
    return reversed_sen

input_sen = input("Enter a sentence: ")
reversed_sen = reverse_words(input_sen)
print("Reversed sentence:", reversed_sen)
```

```python
# 22. Implement a function to find the first non-repeating character in a string.
def find_first_non_repeating_char(string):
    char_count = {}
    
    for char in string:
        char_count[char] = char_count.get(char, 0) + 1

    for char in string:
        if char_count[char] == 1:
            return char


input_string = input("Enter a string: ")
first_non_repeating_char = find_first_non_repeating_char(input_string)
if first_non_repeating_char:
    print("First non-repeating character:", first_non_repeating_char)
else:
    print("No non-repeating character found")


```
